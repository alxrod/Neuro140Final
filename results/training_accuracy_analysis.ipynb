```python
import imageio
import os
import numpy as np
import pandas as pd

import matplotlib.pyplot as plt
from skimage.measure import label
from skimage.io import imread

import pickle as pkl
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

from skimage import data
from skimage.filters import threshold_otsu
from skimage.segmentation import clear_border
from skimage.measure import label, regionprops
from skimage.morphology import closing, square
from skimage.color import label2rgb
```

```python
csvs = [x for x in os.listdir("./") if ".csv" in x]
```

```python
csvs_20 = [x for x in os.listdir("./") if "_epochs.csv" in x]
```

```python
all_dfs = []
for csv in csvs:
    all_dfs.append(pd.read_csv(csv))
all_20_dfs = []
for csv in csvs_20:
    all_20_dfs.append(pd.read_csv(csv))
```

```python
# , label = "line 1", linestyle="-"
for df in all_dfs:
    plt.plot(df["epoch"], df["iou"])
plt.legend(["Body=1, Border=0, Bg=-1","Body=1, Border=1 Bg=0","Body=1, Border=0 Bg=0","Body=1, Border=-1 Bg=0"])
plt.xlabel("Epoch #")
plt.ylabel("IOU")
plt.title("IOU over Training for Different GT Values")
```

```python
# , label = "line 1", linestyle="-"
for df in all_20_dfs:
    plt.plot(df["epoch"], df["iou"])
plt.legend(["Body=1, Border=0 Bg=0","Body=1, Border=-1 Bg=0"])
plt.xlabel("Epoch #")
plt.ylabel("IOU")
plt.title("IOU over Training for Different GT Values")
```

```python
for df in all_dfs[1:]:
    plt.plot(df["epoch"], df["loss"])
plt.legend(["Body=1, Border=1 Bg=0","Body=1, Border=0 Bg=0","Body=1, Border=-1 Bg=0"])
plt.xlabel("Epoch #")
plt.ylabel("Loss")
plt.title("Loss over Training for Different GT Values")
```
